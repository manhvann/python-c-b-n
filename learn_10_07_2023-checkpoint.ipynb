```python
#Cách sử dụng cmp_to_key trong functools
from math import *
from functools import cmp_to_key

#return -1 1 0
#Nếu a là thằng đứng trước và b là thằng đứng sau khi mà bạ sắp xếp
#Nếu a và b đã đứng đúng vị trí thì trả về -1
#Nếu a và b đứng sai vị trí thì trả về 1
# def cmp(a,b):
#     if a < b:
#         return -1
    
#     else:
#         return 1

# a = [10,2,1,5,1,4,6,0]
# a.sort(key = cmp_to_key(cmp))
# print(a)

#Sắp xếp các phần tử trong mảng theo tổng chữ số tăng dần 
#nếu hai số có tổng cs bằng nhau thì số bé đứng trước
def tong(n):
    sum = 0
    while n != 0:
        sum += n % 10
        n //= 10
    return sum

def cmp(a,b):
    tong1,tong2 = tong(a),tong(b)
    if tong1 != tong2:
        return tong1-tong2
    else:
        return a-b

a = [30,3000000,12,21,111,9,19,10,20,39]
a.sort(key = cmp_to_key(cmp))
print(a)
```

```python
#Tìm kiếm nhị phân
def TKNP(a,k):
    l = 0
    r = len(a)-1
    while(l <= r):
        g = (l+r) // 2
        if(a[g] == k):
            return g
        elif a[g] > k:
            r = g-1
        else:
            l = g+1
    return -1
a = [1,2,3,4,5,6,7,8,9]
print("Vi tri la: ",TKNP(a,7))


#Các biến thể
#Tìm vị trí xuất hiện đầu tiên của phần tử
def first_pos(a,l,r,k):
    res = -1
    while(l <= r):
        g = (l+r) // 2
        if(a[g] == k):
            res = g
            r = g-1
        elif a[g] > k:
            r = g-1
        else:
            l = g+1
    return res
a = [1,2,2,3,2,4,2,5,6,7,8,9]
print("Firts pos: ",first_pos(a,0,len(a)-1,2))


#Tìm phần tử đầu tiên có giá trị lớn hơn hoặc bằng x
def first_pos1(a,l,r,k):
    res = -1
    while(l <= r):
        g = (l+r) // 2
        if(a[g] >= k):
            res = g
            r = g-1
        elif a[g] < k:
            l = g + 1
    return res
a = [1,2,2,3,2,4,2,5,5,6,7,8,9]
print("Firts pos1: ",first_pos1(a,0,len(a)-1,5))
```

- Tính chất:
    + Không có thứ tự

    + Set không thể truy cập phần tử thông qua chỉ số
    + Set có thể thay đổi được
    + Set không thể lưu những phần tử có thể thay đổi được ví dụ như list
- Tạo set: Trong dấu đóng mở ngoặc nhọn{}


```python
if __name__ == '__main__':
    s = {1,2,3,4,1,2,3,4,5}
    print(len(s))
    print(s)
    
#     s = {[1,2],"python","java"} #Lối
#     print(s)

    #Tạo set từ list
    a = [1,2,3,4,5,6,1,2,3,5,8,4]
    b = set(a)
    print(b)
    
    #Phương thức thêm một phần tử vào trong set
    s.add(5)
    s.add(10)
    print(s)
    
    #Thêm nhiều phần tử vào trong set 
    s.update(["python","java","C++",12])
    print(s)
    
    
    
    #xóa phần tử trong set nếu phần tử đó không có thì sẽ báo lỗi
    s.remove("python")
    print(s)
    
    s.discard("java")
    print(s)
    
    #Xóa một phần tử bât kì trong set
    s.pop()
    print(s)
    
    #Xóa toàn bộ set
    s.clear()
    print(s)
    
    
    #duyệt set 
    for x in b: #Toán tử in trong set chỉ mất O(1)
        print(x)
        
        
    #Các phép toán trên tập hợp
    s = {"Manh","abc","python"}
    t = {"Manh","amazon","python"}
    
    #Phép hợp
    print("Phép hợp")
    u = s.union(t)
    v = s | t
    print(s)
    print(t)
    print(u)
    print(v)
    
    #Phép giao
    print("Phép giao")
    u = s.intersection(t)
    v = s & t
    print(s)
    print(t)
    print(u)
    print(v)
    
    #Phép Trừ
    print("Phép trừ")
    u = s.difference(t)
    v = s-t
    print(s)
    print(t)
    print(u)
    print(v)
    
    #Phép lấy những phần thuộc 1 trong 2 tập mà bỏ đi phần thuộc cả 2 tập
    print("Phép hợp - Phép giao")
    u = s.symmetric_difference(t)
    v = s ^ t
    print(s)
    print(t)
    print(u)
    print(v)
    
    #Một số phương thức của set
    print("Tập s và tập t  không có phần tử chung: ",end = "")
    print(s.isdisjoint(t)) #Trả về true nếu hai tập ko có phần tử chúng
    
    
    s = {"Manh","python"}
    t = {"Manh","amazon","python"}
    print(s.issubset(t)) #Kiểm tra tập s có là con của tập t ko
    
    
    print(t.issuperset(s)) #Kiểm tra tập t có phải tập cha của tập s ko
    
```

```python
#Đếm các phần tử là số nguyên tố trên đường chéo chính,phụ mỗi phần tử chỉ đếm một lần
from math import *
def nt(n):
    if n<2:
        return False
    for i in range(2, isqrt(n)+1):
        if n % i == 0:
            return False
    return True
if __name__ == '__main__':
    n, m = map(int, input().split())
    a = []
    for i in range(n):
        b =list(map(int,input().split()))
        a.append(b)
    d = {}
    for i in range(n):
            if(nt(a[i][i])):
                if a[i][i] not in d:
                    d[a[i][i]] = 1
                else:
                    d[a[i][i]] += 1
            if(nt(a[i][n-i-1])):
                if a[i][n-i-1] not in d:
                    d[a[i][n-i-1]] = 1
                else:
                    d[a[i][n-i-1]] += 1
    print(len(d))
    
```
